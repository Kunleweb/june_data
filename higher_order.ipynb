Higher Order Functions

--- Functions in python are know as first class citizens -----
1) can be stored in a variable 
2) Can be stored in a data structure
3) are an object
4) can be passed as an argument to another function
5) return function from another function 


There a higher order function takes a function as an argument or returns a function 


Map, Filter and Reduce 

```python
#map; takes a function and applies it to every element of an iterable
def square(x):
    return x*2

a= [1,2,3,4,4]
#tuple(map(square,a ))

#map returns a  map object so we need to use a list or tuple to return the iterables


#Filter: apply a function on an iterable and returns based on condition specified.
def isEven(x):
    return x%2 ==0
#list(filter(isEven, [1,2,3,4,5]))


#Reduce: apply; reduces iterable to a single element and we provid the logic for it 
from functools import reduce
reduce(lambda a,b: a+b, [1,2,3,4,5,6,7]) 
```
